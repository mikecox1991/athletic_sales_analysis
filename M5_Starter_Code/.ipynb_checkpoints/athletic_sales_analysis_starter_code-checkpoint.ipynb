```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df=pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df=pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([sales_2020_df, sales_2021_df], ignore_index=True)
combined_df.reset_index(drop=True, inplace=True)
combined_df.head(20)
```

```python
# Check if any values are null.
null_sum=combined_df.isnull().sum()
null_sum
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], yearfirst=True)
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

total_products_sold_usg = combined_df.groupby(['region', 'state', 'city']).agg(Total_Products_Sold=('units_sold', 'sum'))
total_products_sold_usg.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
top_5_results_usg = total_products_sold_usg.sort_values(by='Total_Products_Sold', ascending=False)
top_5_results_usg.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold_usp= pd.pivot_table(combined_df,
                                        index = ['region', 'state', 'city'],
                                        values = 'units_sold',
                                        aggfunc = 'sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold_usp.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
  
top_5_results_usp = total_products_sold_usp.sort_values(by='Total_Products_Sold', ascending=False)
top_5_results_usp.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_tsg_df=combined_df.groupby(['region', 'state', 'city']).agg(Total_Sales=('total_sales', 'sum'))
total_sales_tsg_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
Top_5_results_tsg= total_sales_tsg_df.sort_values(by='Total_Sales', ascending=False).head()
Top_5_results_tsg
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_tsp_df= pd.pivot_table(combined_df,
                                   index = ['region', 'state', 'city'],
                                   values = 'total_sales',
                                   aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_tsp_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.
Top_5_results_tsp= total_sales_tsp_df.sort_values(by='Total_Sales', ascending=False).head()
Top_5_results_tsp
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_retailerg_df=combined_df.groupby(['retailer', 'region', 'state', 'city']).agg(Total_Sales=('total_sales', 'sum'))
total_sales_retailerg_df.rename(columns={"total_sales" : "Total_Sales"}, inplace=True)

# Show the top 5 results.
Top_5_results_retailg= total_sales_retailerg_df.sort_values(by='Total_Sales', ascending=False).head()
Top_5_results_retailg
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_retailerp_df=pd.pivot_table(combined_df,
                                        index = ['retailer', 'region', 'state', 'city'],
                                        values = 'total_sales',
                                        aggfunc = 'sum')
                                        
# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
Top_5_results_retailp= total_sales_retailerp_df.sort_values(by='total_sales', ascending=False).head()
Top_5_results_retailp
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_shoe_df=combined_df.loc[(combined_df["product"] == "Women's Athletic Footwear")]
womens_shoe_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_units_sold = womens_shoe_df.groupby(['retailer', 'region', 'state', 'city']).agg(Womens_Footwear_Units_Sold = ('units_sold', 'sum'))
womens_units_sold.rename(columns={'units sold' : 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
top_5_womens_units=womens_units_sold.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
top_5_womens_units
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_units_soldp = pd.pivot_table(womens_shoe_df,
                                    index = ['retailer', 'region', 'state', 'city'],
                                    values = 'units_sold',
                                    aggfunc = 'sum')
                                
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_units_soldp.rename(columns={'units_sold' : 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
top_5_womens_unitsp=womens_units_soldp.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
top_5_womens_unitsp
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_day_units = pd.pivot_table(womens_shoe_df,
                                  index = ['invoice_date'],
                                  values = 'total_sales',
                                  aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_day_units.rename(columns={'total_sales' : 'Total Sales'})

# Show the table.
womens_day_units
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_day_units.resample('D').sum()
# Sort the resampled pivot table in ascending order on "Total Sales".
womens_day_units.sort_values(by='total_sales', ascending= False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_day_units.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_day_units.sort_values(by='total_sales', ascending=False).head(10)
```
